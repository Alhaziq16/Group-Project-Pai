# Data Engineer – Cat Breed Classification Dataset

Domain: Animal Subspecies

Objective:
To collect, prepare, standardize, and split the dataset for the classification of 10 cat breeds.

```python
classes = [
    "Persian Cat",
    "Siamese Cat",
    "Maine Coon",
    "Bengal Cat",
    "Ragdoll Cat",
    "British Shorthair",
    "Sphynx Cat",
    "Russian Blue Cat",
    "Scottish Fold",
    "Abyssinian Cat"
]

print("Total classes:", len(classes))
print(classes)
```

```python
for animal_class in classes:

    folder_name = f"dataset/{animal_class.replace(' ', '_')}"

    os.makedirs(folder_name, exist_ok=True)

    print(f"\nDownloading {animal_class}...")

    crawler = BingImageCrawler(
        storage={'root_dir': folder_name}
    )

    crawler.crawl(
        keyword=f"{animal_class} cat photo",
        filters={'type': 'photo'},
        max_num=10,
        overwrite=False
    )

print("\nDataset collection complete!")
```

```python
import os

selected_classes = [
    "Abyssinian",
    "American Bobtail",
    "American Curl",
    "American Shorthair",
    "American Wirehair",
    "Applehead Siamese",
    "Balinese",
    "Bengal",
    "Birman",
    "Bombay"
]

for breed in selected_classes:
    path = f"cat_dataset/images/{breed}"

    count = len(os.listdir(path))

    print(f"{breed}: {count} images")
```

```python
import os

breeds = os.listdir("cat_dataset/images")

for breed in sorted(breeds):
    count = len(os.listdir(f"cat_dataset/images/{breed}"))
    if count >= 100:
        print(breed, ":", count)
```

```python
import os
import shutil

selected_classes = [
    "Abyssinian",
    "American Bobtail",
    "American Shorthair",
    "Bengal",
    "Birman",
    "Bombay",
    "British Shorthair",
    "Maine Coon",
    "Ragdoll",
    "Siamese"
]

os.makedirs("cat_final_dataset", exist_ok=True)

for breed in selected_classes:
    src = f"cat_dataset/images/{breed}"
    dst = f"cat_final_dataset/{breed}"

    shutil.copytree(src, dst)

print("Final dataset created successfully!")
```

```python
for breed in os.listdir("cat_final_dataset"):
    count = len(os.listdir(f"cat_final_dataset/{breed}"))
    print(breed, ":", count)
```

```python
import splitfolders

splitfolders.ratio(
    "cat_final_dataset",
    output="cat_split_dataset",
    seed=42,
    ratio=(0.7, 0.15, 0.15)
)

print("Dataset split completed!")
```

```python
for folder in ["train", "val", "test"]:
    count = len(os.listdir(f"cat_split_dataset/{folder}/Abyssinian"))
    print(folder, ":", count)
```
