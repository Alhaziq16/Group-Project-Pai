```python
import tensorflow as tf

print("TensorFlow version:", tf.__version__)
```

```python
IMG_SIZE = (224, 224)
BATCH_SIZE = 32

train_dataset = tf.keras.utils.image_dataset_from_directory(
    "/content/cat_split_dataset/train",
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE
)

val_dataset = tf.keras.utils.image_dataset_from_directory(
    "/content/cat_split_dataset/val",
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE
)

test_dataset = tf.keras.utils.image_dataset_from_directory(
    "/content/cat_split_dataset/test",
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    shuffle=False
)
```

```python
print(train_dataset.class_names)
```

```python
print("GPU:", tf.config.list_physical_devices('GPU'))
#kena change runtime kalu dok lamo
```

```python
#part ResNet50

from tensorflow.keras.applications import ResNet50
from tensorflow.keras import layers, models

base_model = ResNet50(
    weights='imagenet',
    include_top=False,
    input_shape=(224, 224, 3)
)

base_model.trainable = False

model = models.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.3),
    layers.Dense(10, activation='softmax')
])

model.summary()
```

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
history = model.fit(
    train_dataset,
    validation_data=val_dataset,
    epochs=5
)
```

```python
test_loss, test_acc = model.evaluate(test_dataset)

print("ResNet50 Test Accuracy:", test_acc)
```

ResNet50 (5 epochs)

#DenseNet21

```python
from tensorflow.keras.applications import DenseNet121
from tensorflow.keras import layers, models

base_model = DenseNet121(
    weights='imagenet',
    include_top=False,
    input_shape=(224, 224, 3)
)

base_model.trainable = False

densenet_model = models.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.3),
    layers.Dense(10, activation='softmax')
])

densenet_model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

densenet_model.summary()
```

```python
densenet_history = densenet_model.fit(
    train_dataset,
    validation_data=val_dataset,
    epochs=50
)
```

```python
test_loss, test_acc = densenet_model.evaluate(test_dataset)

print("DenseNet121 Test Accuracy:", test_acc)
print("DenseNet121 Test Loss:", test_loss)
```

| Model       | Epochs | Test Accuracy |  Test Loss |
| ----------- | -----: | ------------: | ---------: |
| DenseNet121 |     50 |    **42.24%** | **1.6657** |

by using DenseNet121

#MOBILENETV3

```python
from tensorflow.keras.applications import MobileNetV3Large
from tensorflow.keras import layers, models

base_model = MobileNetV3Large(
    weights='imagenet',
    include_top=False,
    input_shape=(224, 224, 3)
)

base_model.trainable = False

mobilenet_model = models.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.3),
    layers.Dense(10, activation='softmax')
])

mobilenet_model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

mobilenet_model.summary()
```

```python
mobilenet_history = mobilenet_model.fit(
    train_dataset,
    validation_data=val_dataset,
    epochs=50
)
```

```python
test_loss, test_acc = mobilenet_model.evaluate(test_dataset)

print("mobilenet_model Test Accuracy:", test_acc)
print("mobilenet_model Test Loss:", test_loss)
```

Conclusion comparison:

```python
#calculate 33

from sklearn.metrics import average_precision_score
from tensorflow.keras.utils import to_categorical
import numpy as np

# Get true labels
y_true = []

for images, labels in test_dataset:
    y_true.extend(labels.numpy())

y_true = np.array(y_true)

# Convert to one-hot encoding
y_true_onehot = to_categorical(y_true, num_classes=10)

print(y_true_onehot.shape)
```
